Q:1 Reverse an Array

```python
arr=[1,2,3,4,5]
print(f"initial array:{arr}")
size=len(arr)
i=0
j=size-1
while i<=j:
    temp=arr[i]
    arr[i]=arr[j]
    arr[j]=temp
    i+=1
    j-=1
print(f"reverse array:{arr}")
```

```python
arr = [4,1,8,2,5]

large = float('-inf')
s_large = float('-inf')

for num in arr:
    if num > large:
        s_large = large
        large = num
    elif num > s_large and num != large:
        s_large = num

if s_large == float('-inf'):
    print("No second largest element")
else:
    print(f"Second largest is: {s_large}")
```

7. Rotate Array by k Positions: Rotate the array to the right by k positions.

```python

nums = [1, 2, 3, 4, 5, 6, 7]
print(f"before:{nums}")
k = 3 # number of times need to rotate

n = len(nums)

temp = [0] * n #like make a n size  vector with value 0

k = k % n   # so that line 7 will not give run time error in case of k > n
new_start = n - k

i = 0

for j in range(new_start, n):
    temp[i] = nums[j]
    i += 1

for j in range(0, new_start):
    temp[i] = nums[j]
    i += 1

nums.clear()
nums.extend(temp)
print(f"after:{nums}")
```

10. Merge Two Sorted Arrays


```python
arr1 = [1, 3, 5, 7]
arr2 = [2, 4, 6, 8,10]
print(f"Array 1: {arr1}")
print(f"Array 2: {arr2}")

merged_array = []
ptr1 = 0
ptr2 = 0

while ptr1 < len(arr1) and ptr2 < len(arr2):
    if arr1[ptr1] < arr2[ptr2]:
        merged_array.append(arr1[ptr1])
        ptr1 += 1
    else:
        merged_array.append(arr2[ptr2])
        ptr2 += 1

while ptr1 < len(arr1):
    merged_array.append(arr1[ptr1])
    ptr1 += 1

while ptr2 < len(arr2):
    merged_array.append(arr2[ptr2])
    ptr2 += 1

print(f"Merged Array: {merged_array}")

```

17. Find the Leader Elements: An element is a leader if it is greater than all elements to its right.


```python
arr = [16, 17, 4, 3, 5, 2]
print(f"Original array: {arr}")

leaders = []
max_right = -1

for i in range(len(arr) - 1, 0, -1):
    if arr[i] >= max_right:
        leaders.append(arr[i])
        max_right = arr[i]

leaders.reverse()
print(f"Leader elements: {leaders}")

```

23. Maximum Sum Subarray (Kadane's Algorithm)


```python
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(f"Original array: {arr}")

max_so_far = arr[0]
current_max = arr[0]

for i in range(1, len(arr)):
    current_max = max(arr[i], current_max + arr[i])
    max_so_far = max(max_so_far, current_max)

print(f"Maximum sum subarray: {max_so_far}")


```

24. Rearrange Array Alternately: Rearrange an array such that elements alternate between the largest and smallest.


```python
arr = [1, 2, 3, 4, 5, 6, 7]
print(f"Original array: {arr}")

arr.sort()
result=[]
i=0
j=len(arr)-1
while i<=j:
  if i==j:
    result.append(arr[i])
    break
  else:
    result.append(arr[j])
    result.append(arr[i])
    i+=1
    j-=1
print(result)
```

26. Find Peak Element: A peak element is greater than its neighbors. Find one such element.


```python
arr=[8,3,9,6,7,2,4]
print(f"Original array: {arr}")
prev=arr[0]
result=[]
for i in range(1,len(arr)-1):
  nxt=arr[i+1]
  if prev<arr[i] and arr[i]>nxt:
    result.append(arr[i])
  prev=arr[i]
print(f"Peak elements: {result}")
```

28. Sort an Array of 0s, 1s, and 2s: Sort an array consisting of only 0s, 1s, and 2s.


```python
arr = [0, 1, 2, 0, 1, 2, 0]
print(f"original array: {arr}")

l = 0
m = 0
h = len(arr) - 1

while m <= h:
    if arr[m] == 0:
        arr[l], arr[m] = arr[m], arr[l]
        l += 1
        m += 1
    elif arr[m] == 1:
        m += 1
    else:
        arr[m], arr[h] = arr[h], arr[m]
        h -= 1
print(f"sorted array: {arr}")


```

29. Find the Longest Consecutive Sequence: Find the length of the longest consecutive sequence of integers.


```python
arr = [100, 4, 200, 1, 3, 2]
print(f"Original array: {arr}")

if not arr:
    longest_streak = 0
else:
    num_set = set(arr)
    longest_streak = 0

    for num in arr:
        if (num - 1) not in num_set:
            current_num = num
            current_streak = 1


            while (current_num + 1) in num_set:
                current_num += 1
                current_streak += 1

            longest_streak = max(longest_streak, current_streak)

print(f"Length of the longest consecutive sequence: {longest_streak}")

```

31. Find Equilibrium Index: Find an index such that sum of     elements on left = sum on right.

```python
arr = [1, 2, 3, 4, 0, 6]
print(f"Original array: {arr}")

total_sum = sum(arr)
left_sum = 0
equi_index = -1

for i in range(len(arr)):
    right_sum = total_sum - left_sum - arr[i]
    if left_sum == right_sum:
        equi_index = i
        break
    left_sum += arr[i]

if equi_index != -1:
    print(f"Equilibrium index is: {equi_index}")
else:
    print("No equilibrium index found.")
```
